```python
import openslide

# .svs   
svs_path = "/mnt/TCGA/Data/4ecf8840-d45e-4be3-9d2a-459a8081d62c/TCGA-BR-4369-01Z-00-DX1.69000607-7c5c-4a97-bdc1-f5a4fdfdd39c.svs"

# OpenSlide  
slide = openslide.OpenSlide(svs_path)

#    ( )
print(" WSI    (Shape):", slide.dimensions)

#     ( )
print("    :")
for i, level in enumerate(slide.level_dimensions):
    print(f"   - Level {i}: {level} ( : {slide.level_downsamples[i]})")

#  
print("\n WSI :")
for key, value in slide.properties.items():
    print(f"   {key}: {value}")

#    (  )
if "aperio.Barcode" in slide.properties:
    print("\n  :", slide.properties["aperio.Barcode"])

# Objective Power ( )
if "tiff.XResolution" in slide.properties:
    print("  (Objective Power):", slide.properties.get("tiff.XResolution"))

# Close the slide object
slide.close()
```
