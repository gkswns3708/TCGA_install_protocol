```python
import openslide

# .svs 파일 경로 설정
svs_path = "/mnt/TCGA/Data/4ecf8840-d45e-4be3-9d2a-459a8081d62c/TCGA-BR-4369-01Z-00-DX1.69000607-7c5c-4a97-bdc1-f5a4fdfdd39c.svs"

# OpenSlide 객체 생성
slide = openslide.OpenSlide(svs_path)

# 전체 이미지 크기 (원본 해상도)
print("🔹 WSI 전체 이미지 크기 (Shape):", slide.dimensions)

# 다중 해상도 레벨 확인 (저해상도 버전)
print("🔹 사용 가능한 해상도 레벨:")
for i, level in enumerate(slide.level_dimensions):
    print(f"   - Level {i}: {level} (다운샘플링 비율: {slide.level_downsamples[i]})")

# 메타데이터 출력
print("\n🔹 WSI 메타데이터:")
for key, value in slide.properties.items():
    print(f"   {key}: {value}")

# 바코드 정보 출력 (일부 파일에서 존재)
if "aperio.Barcode" in slide.properties:
    print("\n🔹 바코드 정보:", slide.properties["aperio.Barcode"])

# Objective Power (배율 정보)
if "tiff.XResolution" in slide.properties:
    print("🔹 배율 (Objective Power):", slide.properties.get("tiff.XResolution"))

# Close the slide object
slide.close()
```
